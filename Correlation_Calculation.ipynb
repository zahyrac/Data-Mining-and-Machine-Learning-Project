```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Display related options
%matplotlib inline
# palette options: "colorblind", 'Set1', "Set2", "Set3", "summer", "spring", "pastel", "rainbow"

sns.set(palette = "rainbow")

pd.set_option('display.precision', 2)
```

```python
#Read dataframe in pandas
behaviour = pd.read_csv("Behavioural Score.csv")
behaviour.head(10)
```

```python
#Calculate correlation between behavioral scores
behavioural_corr = behaviour.corr()
behavioural_corr
```

```python
#Heatmap for correlation between behavioral scores
# cmap is very much like color palette
#sns.heatmap(data=nba_corr , annot = True , cmap ="Set1" )

#annot = annotation to give column names on X and Y axis
#cmap = colour map
sns.heatmap(data=behavioural_corr, annot = True, cmap = "coolwarm");
```

```python
#Pearson Correlation Interpretation
#Values near or = +1: Positively corelated
#Values near 0: No correlation
#Values near or = -1: Negatively correlated


#Correlation between behavioural data between:

#SS and Impulsive = 0.62
#Cscore and Escore = 0.31
#SS and Oscore = 0.42
```

### Demo and Drugs

```python
#Read dataframe in pandas
demographic = pd.read_csv("Demo.csv")
demographic.head(10)
```

```python
#Codification of categorical varibales in Demograohic attributes
demographic["Age"] = pd.Categorical(demographic["Age"])
demographic["Age Code"] = demographic["Age"].cat.codes

demographic["Gender"] = pd.Categorical(demographic["Gender"])
demographic["Gender Code"] = demographic["Gender"].cat.codes

demographic["Education"] = pd.Categorical(demographic["Education"])
demographic["Education Code"] = demographic["Education"].cat.codes

demographic["Country"] = pd.Categorical(demographic["Country"])
demographic["Country Code"] = demographic["Country"].cat.codes

demographic["Ethnicity"] = pd.Categorical(demographic["Ethnicity"])
demographic["Ethnicity Code"] = demographic["Ethnicity"].cat.codes
```

```python
#New dataframe with codified demographic attributes
demographic.to_csv("Codified Demo.csv")
demographic.head()
```

```python
#Cross tabulation for Correlation between categorical and continous variable
pd.crosstab(index=demographic["Gender"],columns=demographic["Cannabis Binary"], normalize="index")
```

```python
#Cross tabulation for Correlation between categorical and continous variable
pd.crosstab(index=demographic["Age"],columns=demographic["Cannabis Binary"], normalize="index")
```

```python
#New dataframe for demogarphic
demographic_drugs = pd.read_csv("Codified Demo.csv")
demographic_drugs.head(10)
```

```python
#Correlation between demographic attributes
demograpphic_and_drug_corr = demographic_drugs.corr()
demograpphic_and_drug_corr
```

```python
#Dataframe with correlation of Demographic and Binary drugs
demograpphic_and_drug_corr.to_csv("Demographic and Drug Analysis.csv", sep=",")
```

```python
demo_new = pd.read_csv("Demographic and Drug Analysis.csv", index_col = "Drugs")
demo_new.head()
```

```python
#Heatmap for correlation between demograohic attributes and drugs
sns.heatmap(data=demo_new, xticklabels=True, yticklabels=True, annot = True, cmap = "coolwarm");
```
